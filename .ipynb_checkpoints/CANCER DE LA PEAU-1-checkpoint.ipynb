## decoupage de l'image en 784 pixels

```python
def trans_pix(im,t_x=784):
    imge=Image.open(im)
    
    w=imge.width
    h=imge.height
    #comptage des lignes et des colonnes
    l=np.int64((h-h%t_x)/t_x)
    c=np.int64((w-w%t_x)/t_x)
    d=np.int64(l*c)
    ## les informations de mon tableau
    print(f'Image:{im}')
    print(f'Dimension : h{h} w{w}')
    print(f'nombre des pixels :d{d}')  
    print(f'nombre de ligne et colonne :l{l} c{c}')  
    r=np.linspace(0,w,l+1)
    r_d_tuple=[(np.int64(r[i]),np.int64(r[i])+d) for i in range(0,len(r)-1)]
    
    q=np.linspace(0,h,c+1)
    q_d_tuple=[(np.int64(q[i]),np.int64(q[i])+d) for i in range(0,len(q)-1)]
    
    
    # remplissage de mon tableau des pixels
    pixels=[]
    for lin in range(l):
        for col in range(c):
            [y1,y2,x1,x2]=*r_d_tuple[lin],*q_d_tuple[col]
            x2=x1+t_x
            y2=y1+t_x
            titre=imge.crop((x1,y1,x2,y2))
            t_coord={'x1':x1,'x2':x2,'y1':y1,'y2':y2}
            pixels.append({'image':titre,'coord':t_coord})
    return pixels,d
```

## decoupage de l'image en tableau

```python
def decoupage(im):
    pixels,dim=trans_pix(im)
    l=0
    #cration des sous fichiers des images
    for k in range(dim):
        pixs=pixels[k]['image']
        pix=np.array(pixs)
        fil=open('cancer/'+f'image{k}.csv','w')
        for i in range(784):
            fil.write(f'pixel000{i}'+',')
        fil.write('\n')
        for i in range(784):
            l=0
            for j in range(784):
                for d in range(3):
                    l+=1
                    fil.write(f'{pix[i][j][d]}'+',')
                    if(l==784):
                        fil.write('\n')
                        l=0
        fil.close()
```

## Regions de decision

```python
def plot_decision_regions(X, y, classifier, test_idx=None,
                          resolution=0.02):
# setup marker generator and color map
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('blue', 'red', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])
    
# plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1 
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1],
                    alpha=0.8, c=colors[idx],
                    marker=markers[idx], label=sortie[cl],
                    edgecolor='black')
# highlight test samples
    if test_idx:
# plot all samples
        X_test, y_test = X[test_idx, :], y[test_idx]
        plt.scatter(X_test[:, 0], X_test[:, 1],
                    c='', edgecolor='black', alpha=1.0,
                    linewidth=1, marker='o',
                    s=100, label='test set') 
```

```python
def vue(model,X,y,ax=None,cmap='rainbow'):
    ax=ax or plt.gca()
    #traage des points d'apprentissage
    plt.scatter(X[:,0],X[:,1],c=y,s=30,cmap=cmap,zorder=3)
    ax.axis('tight')
    ax.axis('off')
    xlim=ax.get_xlim()
    ylim=ax.get_ylim()
    #utilisation de l'estimateur
    """model.fit(X,y)"""
    xx,yy=np.meshgrid(np.linspace(*xlim,num=200),np.linspace(*ylim,num=200))
    Z=model.predict(np.c_[xx.ravel(),yy.ravel()]).reshape(xx.shape)
    #creation de la couleur des points
    n_classes=len(np.unique(y))
    contours=ax.contourf(xx,yy,Z,alpha=0.3,levels=np.arange(n_classes+1)-0.5,cmap=cmap,zorder=1)
    ax.set(xlim=xlim,ylim=ylim) 
```
